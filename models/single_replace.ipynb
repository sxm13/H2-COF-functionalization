```python
data = pd.read_excel("./COF.xlsx",sheet_name='two_add')
```

```python
for i in range(50):
    
#     try:
        structure = Atoms.load("./structure/88/"+data["Name"][i]+".cif")
        cluster = Atoms.load("./structure/group/query.cml")
        replace = Atoms.load("./structure/group/two_replacement.cml")

        structure_re_half = replace_pattern_in_structure(structure, cluster, replace, replace_fraction=0.5, atol=1, ignore_atoms_should_not_be_deleted_twice=True)
        structure_re_half.to_ase().write("./structure/88/replace/"+data["Name"][i]+"_all_0.5.cif")
        
        structure_re_1 = replace_pattern_in_structure(structure, cluster, replace, replace_fraction=1, atol=1, ignore_atoms_should_not_be_deleted_twice=True)
        structure_re_1.to_ase().write("./structure/88/replace/"+data["Name"][i]+"_all_1.0.cif")
        
        print(data["Name"][i],"success")
        
#     except:
#         print(data["Name"][i],"fail")
```

```python
for i in range(50):
    structure1 = open("./structure/88/replace/"+data["Name"][i]+"_all_0.5.cif")
    structure2 = open("./structure/88/replace/"+data["Name"][i]+"_all_1.0.cif")
    
    lines1 = structure1.readlines()
    lines2 = structure2.readlines()
    
    structure1.close()
    structure2.close()
    
    new_lines1 = []
    new_lines2 = []
    
    for line in lines1:
        if 'Te' in line:
            new_line = re.sub(r'Te(?!\d)', 'O', line)
            new_lines1.append(new_line)
        elif 'Sn' in line:
            new_line = re.sub(r'Sn(?!\d)', 'C', line)
            new_lines1.append(new_line)
        else:
            new_lines1.append(line)
            
    for line in lines2:
        if 'Te' in line:
            new_line = re.sub(r'Te(?!\d)', 'O', line)
            new_lines2.append(new_line)
        elif 'Sn' in line:
            new_line = re.sub(r'Sn(?!\d)', 'C', line)
            new_lines2.append(new_line)
        else:
            new_lines2.append(line)
            
    new_file1 = open("./structure/88/structures_2/"+data["Name"][i]+"_all_0.5.cif", 'w')
    new_file1.writelines(new_lines1)
    new_file1.close()
    
    new_file2 = open("./structure/88/structures_2/"+data["Name"][i]+"_all_1.0.cif", 'w')
    new_file2.writelines(new_lines2)
    new_file2.close()
    
    print(data["Name"][i],"success")
```
